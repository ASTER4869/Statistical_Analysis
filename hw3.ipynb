```python
import pandas as pd
import os
```

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
df=df.drop_duplicates()
df=df.dropna()
df
##去除空白值和重复值
```

```python

df_zscore = df.copy()  # 复制一个用来存储Z-score得分的数据框
cols = df.columns  #  获得列表框的列名
for col in cols:
    df_col = df[col]  #  得到每一列的值
    if(col!='SMILES'):
        z_score =(df_col - df_col.mean()) / df_col.std() #  计算每一列的Z-score得分
        df_zscore[col] = z_score.abs() > 2.2  # 判断Z-score得分是否大于2.2，如果是则是True，否则为False
    else:
        df_zscore[col]=False
for col in cols:
    df_drop_outlier = df[df_zscore[col] == False]
##处理异常值
df=df_drop_outlier
df
```

```python
df_Spearman=df.corr(method='spearman')['pIC50'].abs().sort_values(ascending=False)
df_Spearman[:20]
df_Spearman[1:20]
```

```python
df_x=list(df_Spearman[1:20].keys())
df_x
```

```python
cols = df.columns  #  获得列表框的列名
X = df.copy()
for col in cols:
    if(col not in df_x):
        X = X.drop([col],axis=1) # 特征
y = df.loc[:, ['pIC50']] # 目标
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
y_pIC50_train = y_train.loc[:,"pIC50"]
y_pIC50_test = y_test.loc[:,"pIC50"]
```

```python
#使用随机森林的方法构建模型
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators = 500,min_samples_leaf = 1,max_depth=40)
rf_model.fit(X_train,y_pIC50_train) # 预测IC50_nM的模型训练
```

```python
score = rf_model.score(X_test, y_test)
score
```

```python

from sklearn.model_selection import GridSearchCV
#GridSearch网格搜索 进行参数调优
rfc=RandomForestRegressor()
param = {"n_estimators": range(1,500),"min_samples_leaf": range(1,3)}     #要调优的参数
gs = GridSearchCV(estimator=rfc,param_grid=param,cv=5)
gs.fit(X_train,y_pIC50_train)

#导出调参后最优参数
best_score=gs.best_score_
best_params=gs.best_params_
print(best_score,best_params,end='\n')
```
